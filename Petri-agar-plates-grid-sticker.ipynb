```
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Define the parameters
initial_x = 0
initial_y = 0
square_size_mm = 6
row_height = 6
adjusted_font_size = 10
rows = [2, 6, 8, 10, 10, 12, 12, 12, 10, 8, 6, 4]
row_offsets = [30.0, 18.0, 12.0, 6.0, 6.0, 0.0, 0.0, 0.0, 6.0, 12.0, 18.0, 24.0]

fig, ax = plt.subplots(figsize=(10, 8))

# Calculate the total number of squares in each row to adjust the numbering from right to left
total_numbers_per_row = [sum(1 for _ in range(num_squares)) for num_squares in rows]

square_number = 1

current_y = initial_y  # Start from the top again
for i, (num_squares, offset) in enumerate(zip(rows, row_offsets)):
    current_x = initial_x + offset
    # For right to left numbering, calculate the starting square number for the current row
    starting_number_for_row = square_number + total_numbers_per_row[i] - 1
    for _ in range(num_squares):
        # Draw square
        ax.add_patch(patches.Rectangle((current_x, current_y), square_size_mm, square_size_mm, edgecolor='black', facecolor='none'))
        # Place the number very close to the upper left corner of the square, maintaining the 0.5 mm offset
        text_x = current_x + 0.5  # 0.5 mm from the left edge
        text_y = current_y + row_height - 0.5  # 0.5 mm from the top edge
        # Use starting number for row and decrement for each square to number from right to left
        ax.text(text_x, text_y, str(starting_number_for_row), ha='left', va='top', fontsize=adjusted_font_size)
        # Move to the next square position in the row
        current_x += square_size_mm
        # Decrement the square number for right to left numbering
        starting_number_for_row -= 1
    # Increment the total square number by the number of squares in the current row for the next row's calculation
    square_number += num_squares
    # Move down to the next row
    current_y -= row_height

ax.set_aspect('equal', 'box')
ax.autoscale_view()
ax.axis('off')

plt.show()
```

```
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Define the parameters
initial_x = 0
initial_y = 0
square_size_mm = 6
row_height_mm = 6
adjusted_font_size = 6  # Reduced font size for smaller numbers
rows = [2, 6, 8, 10, 10, 12, 12, 12, 10, 8, 6, 4]
row_offsets_mm = [30.0, 18.0, 12.0, 6.0, 6.0, 0.0, 0.0, 0.0, 6.0, 12.0, 18.0, 24.0]

# Convert square size and offsets to inches for matplotlib (1 inch = 25.4 mm)
square_size_in = square_size_mm / 25.4
row_height_in = row_height_mm / 25.4
row_offsets_in = [offset / 25.4 for offset in row_offsets_mm]

# Create figure with appropriate size in inches and high DPI for printing
fig, ax = plt.subplots(figsize=(10, 8), dpi=300)  # Adjust figsize as needed based on actual layout

# Adjusting linewidth for a finer grid
line_width = 0.1  # Finer line width for the grid

current_y = initial_y  # Start from the top again
for i, (num_squares, offset_in) in enumerate(zip(rows, row_offsets_in)):
    current_x = initial_x + offset_in
    starting_number_for_row = sum(rows[:i+1])
    for _ in range(num_squares):
        ax.add_patch(patches.Rectangle((current_x, current_y), square_size_in, square_size_in, edgecolor='black', facecolor='none', linewidth=line_width))
        # Adjust text position for upper left corner; consider a slight offset from the edge
        text_x = current_x + (0.5 / 25.4)  # Slight offset from the left edge in inches
        text_y = current_y + square_size_in - (0.5 / 25.4)  # Slight offset from the top edge in inches
        ax.text(text_x, text_y, str(starting_number_for_row), ha='left', va='top', fontsize=adjusted_font_size)
        current_x += square_size_in
        starting_number_for_row -= 1
    current_y -= row_height_in

ax.set_aspect('equal', 'box')
ax.autoscale_view()
ax.axis('off')

# Save to a high-resolution PDF file
plt.savefig("squares_grid_adjusted.pdf", dpi=300, format='pdf', bbox_inches='tight')
```

With Colab you can harness the full power of popular Python libraries to analyze and visualize data. The code cell below uses **numpy** to generate some random data, and uses **matplotlib** to visualize it. To edit the code, just click the cell and start editing.
